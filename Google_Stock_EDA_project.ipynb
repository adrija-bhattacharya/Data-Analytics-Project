```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#loading the dataset
df = pd.read_csv(r"C:\Users\adrija\Downloads\GOOG.csv")

df.head()
```

```python
# Seeing the shape in a row N column 
print("Shape of the data:", df.shape
df.info()
# Also checking if anything is missing
print("\nMissing values:\n")
print(df.isnull().sum())
```

```python
# changing the date column using date=time function of pandas
df['Date'] = pd.to_datetime(df['Date'])

# I think it is better if I make 'Date' the index of the table
df.set_index('Date', inplace=True)

# Let me check the cleaned-up data
df.head()
```

```python
# average, min, max, etc.
df.describe()
```

```python
#data visualisation
#using line plot for the trend of closing price over time
plt.figure(figsize=(12,6))
plt.plot(df['Close'], label='Close Price', color='blue')
plt.title("Google's Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#usinf line plot for checking the volume of trading stock over time
plt.figure(figsize=(12,4))
plt.plot(df['Volume'], color='purple')
plt.title("Google's Trading Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

```python
#checking the correlation among the data columns
plt.figure(figsize=(8,6))
sns.heatmap(df[['Open', 'High', 'Low', 'Close', 'Volume']].corr(), annot=True, cmap='YlGnBu')
plt.title("Correlation Between Different Stock Metrics")
plt.show()
```

Exploration observations:
The stock price increased a lot over time.
increase in trading volume.
All prices are strongly correlated.
Generally we use heatmaps to find correlatiom between many different types of stocks.We observe a higher relation between open,high,low and close from this data set.
