**catagorizing imdb_score column
we will reduce the imdb_score values ito 4 catagories:Popular,average,below_average,not_popular**

```python
def categorize_col(df, cols, labels):
    for col in cols:
        edges = [
            df[col].describe()['min'],
            df[col].describe()['25%'],
            df[col].describe()['50%'],
            df[col].describe()['75%'],
            df[col].describe()['max']
        ]
        df[col] = pd.cut(df[col], edges, labels=labels, duplicates='drop')
    return df

# Correct function usage:
labels = ['Not popular', 'Below_average', 'Average', 'Popular']
df = categorize_col(df, ['tmdb_popularity'], labels)
print(df['tmdb_popularity'].unique())
```
